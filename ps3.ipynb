```python
import pylab as py
import numpy as np
import matplotlib.pyplot as plt
```

```python
def main():
    t = np.linspace(0, 60, 100)
    #initial perturbation z(0)=1, renv
    renv1=6 #k/km
    renv2=7
    renv3=8
    rdry=9.8
    T1=15+273.15
    T2=25+273.15
    T3=35+273.15
    g=0.0098 #km/s2
    #buoyancy frequency
    N1a=np.sqrt(((rdry-renv1)*g)/T1)
    N1b=np.sqrt(((rdry-renv1)*g)/T2)
    N1c=np.sqrt(((rdry-renv1)*g)/T3)
    N2a=np.sqrt(((rdry-renv2)*g)/T1)
    N2b=np.sqrt(((rdry-renv2)*g)/T2)
    N2c=np.sqrt(((rdry-renv2)*g)/T3)
    N3a=np.sqrt(((rdry-renv3)*g)/T1)
    N3b=np.sqrt(((rdry-renv3)*g)/T2)
    N3c=np.sqrt(((rdry-renv3)*g)/T3)
    print (N1a)
    print (N1b)
    print (N1c)
    print (N2a)
    print (N2b)
    print (N2c) 
    print (N3a)
    print (N3b)
    print (N3c)
    z1a= 1*np.cos(N1a*t) 
    z1b= 1*np.cos(N1b*t)
    z1c= 1*np.cos(N1c*t)
    z2a= 1*np.cos(N2a*t)
    z2b= 1*np.cos(N2b*t)
    z2c= 1*np.cos(N2c*t)
    z3a= 1*np.cos(N3a*t)
    z3b= 1*np.cos(N3b*t)
    z3c= 1*np.cos(N3c*t)
        
    plt.figure()
    plt.plot(t, z1a,'m',label='renv=6K/km;T=15')
    plt.plot(t, z1b,'c',label='renv=6K/km;T=25')
    plt.plot(t, z1c,'b',label='renv=6K/km;T=35')
    plt.plot(t, z2a,'m',label='renv=7K/km;T=15')
    plt.plot(t, z2b,'c',label='renv=7K/km;T=25')
    plt.plot(t, z2c,'b',label='renv=7K/km;T=35')
    plt.plot(t, z3a,'m',label='renv=8K/km;T=15')
    plt.plot(t, z3b,'c',label='renv=8K/km;T=25')
    plt.plot(t, z3c,'b',label='renv=8K/km;T=35')
    plt.xlabel('$Time\ (minutes)$',fontsize=18)
    #py.xlim([250,300])
    #py.ylim([1,100])
    plt.ylabel('$Air\ parcel\ perturbation\ z\ (m))$',fontsize=18)
    plt.yscale('log')
    plt.grid(True)

    plt.legend(loc='upper right')
    plt.rcParams['figure.figsize']=(10,10)
    plt.savefig('ps3-airparcelperturbation.png')
   

    plt.show()

if __name__ == '__main__':
    main()
```
