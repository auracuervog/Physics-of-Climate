```python
import pylab as py
import numpy as np
import matplotlib.pyplot as plt
```

```python
def main():
    x = np.linspace(250, 300, 100)
    es = 6.11*np.exp((2230000/461.6)*((1/273.15)-(1/x)))
    
    plt.figure()
    plt.plot(x, es,'g',label='es')
    plt.xlabel('$T(K)$',fontsize=18)
    py.xlim([250,300])
    py.ylim([1,100])
    plt.ylabel('$Saturation water vapor pressure(hPa)$',fontsize=18)
    plt.yscale('log')
    plt.grid(True)

    plt.legend(loc='upper right')
    plt.rcParams['figure.figsize']=(10,10)
    plt.savefig('ps2ees436f.png')
   
    plt.show()

if __name__ == '__main__':
    main()
```
