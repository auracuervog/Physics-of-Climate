```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def main():
    z = np.linspace(0, 40, 100)
    x1 = 1000*np.exp(-z/7)
    x2 = 1000*np.exp(-z/6)
    x3 = 1000*np.exp(-z/8)

    plt.figure()
    plt.plot(x1, z,'r',label='Scale height = 7 km')
    plt.plot(x2, z, 'b',label='Scale height = 6 km')
    plt.plot(x3,z,'g', label='Scale height = 8 km')
    plt.xlabel('$P(mb)$')
    plt.ylabel('$z(km)$')
    plt.legend(loc='upper right')
    plt.rcParams['figure.figsize']=(10,10)
    plt.savefig('ps1ees436.png')
   
    plt.show()

if __name__ == '__main__':
    main()
```
